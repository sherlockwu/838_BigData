```python
import py_entitymatching as em
import pandas as pd
import logging
```

```python
table_a = em.read_csv_metadata('anime_1.csv', key='ID')
table_b = em.read_csv_metadata('animePlayer.csv', key='ID')
```

```python
title_blocker = em.OverlapBlocker()
title_blocker_result = title_blocker.block_tables(table_a,table_b,'Title','Title',q_val=4 , word_level=False,overlap_size=7,
                                                 l_output_attrs=['Title','Genres_Merged','Episodes','Year','Type','ProductionHouse_Merged'],
                                                  r_output_attrs=['Title','Genres_Merged','Episodes','Year','Type','ProductionHouse_Merged'])
title_blocker_result
```

```python
corres = em.get_attr_corres(table_a,table_b)
```

```python
title_debug = em.debug_blocker(title_blocker_result,table_a,table_b,attr_corres=corres['corres'], output_size=2000)
```
