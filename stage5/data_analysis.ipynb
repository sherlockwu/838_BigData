<H5>Step 1 : Load the dataset</H5>

```python
from sqlalchemy import create_engine
import sqlite3
import os
#if the database file exists, then delte the file
path = 'Olapdatabase.db'
if (os.path.isfile(path)):
    os.remove(path)
# create a database where we'll load the dataset from the csv file
db_conn = create_engine('sqlite:///Olapdatabase.db')
```

```python
import string
# load in a batch of 50000 tuples. Modify this value to your needs
chunks = 500
for data in pd.read_csv(data_file,chunksize=chunks,
iterator=True, encoding='utf-8'):
    data = data.rename(columns={col: col.replace('-', ' ') for col in data.columns})
    data = data.rename(columns={col: col.strip() for col in data.columns})
    data = data.rename(columns={col: string.capwords(col) for col in data.columns})
    data = data.rename(columns={col: col.replace(' ', '') for col in data.columns})
    data.to_sql('data', db_conn.raw_connection(), flavor='sqlite', if_exists='append')
```

##### <H5>Data Exploration Case 5</H5>   <body>     <p> Get the rating of different types of animes </p>     </body>

```python
movie_rating = pd.read_sql("""SELECT Rating FROM data
where Type like "%Movie%" and typeof(Rating) = "real"
""", db_conn.raw_connection())

tv_rating = pd.read_sql("""SELECT Rating FROM data
where Type like "%TV%" and typeof(Rating) = "real"
""", db_conn.raw_connection())

ova_rating = pd.read_sql("""SELECT Rating FROM data
where Type like "%OVA%" and typeof(Rating) = "real"
""", db_conn.raw_connection())

tvspe_rating = pd.read_sql("""SELECT Rating FROM data
where Type like "%TV Special%" and typeof(Rating) = "real"
""", db_conn.raw_connection())

web_rating = pd.read_sql("""SELECT Rating FROM data
where Type like "%Web%" and typeof(Rating) = "real"
""", db_conn.raw_connection())

music_rating = pd.read_sql("""SELECT Rating FROM data
where Type like "%Music Video%" and typeof(Rating) = "real"
""", db_conn.raw_connection())

dvdspe_rating = pd.read_sql("""SELECT Rating FROM data
where Type like "%DVD Special%" and typeof(Rating) = "real"
""", db_conn.raw_connection())

other_rating = pd.read_sql("""SELECT Rating FROM data
where Type like "%Other%" and typeof(Rating) = "real"
""", db_conn.raw_connection())

display(movie_rating)
```

##### <H5>Data Visualization Case 1</H5>     <body>     <p>Plot the average AQI of Madison and New York each year.</p>     </body>

```python
## numpy is used for creating fake data
import numpy as np 
import matplotlib as mpl 

## agg backend is used to create plot as a .png file
#mpl.use('agg')

import matplotlib.pyplot as plt 
## Create x label
numBoxes = 8
randomDists = ['Movie', 'TV', 'OVA', 'TV Special',
               'Web', 'Music', 'DVD Special', 'Other']
movie_rating_l = list(movie_rating['Rating'].values.flatten())
tv_rating_l = list(tv_rating['Rating'].values.flatten())
ova_rating_l = list(ova_rating['Rating'].values.flatten())
tvspe_rating_l = list(tvspe_rating['Rating'].values.flatten())
web_rating_l = list(web_rating['Rating'].values.flatten())
music_rating_l = list(music_rating['Rating'].values.flatten())
dvdspe_rating_l = list(dvdspe_rating['Rating'].values.flatten())
other_rating_l = list(other_rating['Rating'].values.flatten())

data = [movie_rating_l, tv_rating_l, ova_rating_l, tvspe_rating_l, web_rating_l, music_rating_l, dvdspe_rating_l, other_rating_l]
# basic plot
plt.boxplot(data, notch=0, sym='+', vert=1, whis=1.5)

fig, ax1 = plt.subplots(figsize=(10, 6))
fig.canvas.set_window_title('Boxplot')
plt.subplots_adjust(left=0.075, right=0.95, top=0.9, bottom=0.25)

bp = plt.boxplot(data, notch=0, sym='+', vert=1, whis=1.5)
plt.setp(bp['boxes'], color='black')
plt.setp(bp['whiskers'], color='black')
plt.setp(bp['fliers'], color='red', marker='+')

# Add a horizontal grid to the plot, but make it very light in color
# so we can use it for reading data values but not be distracting
ax1.yaxis.grid(True, linestyle='-', which='major', color='lightgrey',
               alpha=0.5)

# Hide these grid behind plot objects
ax1.set_axisbelow(True)
ax1.set_title('Comparison of animes\' rating based on different types')
ax1.set_xlabel('Anime Types')
ax1.set_ylabel('Rating')

# Set the axes ranges and axes labels
ax1.set_xlim(0.5, numBoxes + 0.5)
top = 5
bottom = 0
ax1.set_ylim(bottom, top)
xtickNames = plt.setp(ax1, xticklabels=np.repeat(randomDists, 2))
plt.setp(xtickNames, rotation=45, fontsize=8)

# Save the figure
fig.savefig('fig1.png', bbox_inches='tight')
plt.show()
```
