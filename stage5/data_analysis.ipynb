```python
import pandas as pd
# numpy has a lots of useful math related modules
import numpy as np
# Helpful function to display intermittent result
from IPython.display import display
```

<H4> Step 0 : Take a peek at the dataset</H4>

```python
# Load the data and display only first five rows
# NOTE replace the following value with the actual path to the csv file
data_file = "./AQI.csv" # e.g /u/u/p/..../cs564/p5/AQI.csv"
display(pd.read_csv(data_file, nrows=5).head())
```

<H5>Step 1 : Load the dataset</H5>

```python
from sqlalchemy import create_engine
import sqlite3
import os
#if the database file exists, then delte the file
path = 'AirQualityIndex.db'
if (os.path.isfile(path)):
    os.remove(path)
# create a database where we'll load the dataset from the csv file
db_conn = create_engine('sqlite:///AirQualityIndex.db')
```

```python
import string
# load in a batch of 50000 tuples. Modify this value to your needs
chunks = 50000
for data in pd.read_csv(data_file,chunksize=chunks,
iterator=True, encoding='utf-8'):
    data = data.rename(columns={col: col.replace('-', ' ') for col in data.columns})
    data = data.rename(columns={col: col.strip() for col in data.columns})
    data = data.rename(columns={col: string.capwords(col) for col in data.columns})
    data = data.rename(columns={col: col.replace(' ', '') for col in data.columns})
    data.to_sql('data', db_conn.raw_connection(), flavor='sqlite', if_exists='append')
```

##### <H5>Step 2 : Data Exploration</H5>

##### <H5>Data Exploration Case 1</H5>   <body>     <p> Get the total number of tuples in the database </p>     </body>

```python
total_rows = pd.read_sql("""SELECT count(*) as Total_Rows FROM data
 """, db_conn.raw_connection())
# Display the result (Note - the result 'avg_air_quality' is an instance of 'pandas DataFrame
display(total_rows)
```

###### <H5>Data Exploration Case 2</H5>   <body>     <p> Find the average air quality index for each State all the years</p>     </body>

```python
avg_air_quality = pd.read_sql("""SELECT AVG(AQI) as avg_aqi,
 StateName as state FROM data
 GROUP BY state
 ORDER BY avg_aqi DESC""", db_conn.raw_connection())
# Display the result (Note - the result 'avg_air_quality' is an instance of 'pandas DataFrame
display(avg_air_quality)
```

##### <H5>Data Exploration Case 3</H5>   <body>     <p> Find the first ten cities with the hightest average air quality</p>     </body>

```python
first_ten_high = pd.read_sql("""SELECT AVG(AQI) as avg_aqi,
 CityName as city, StateName as state FROM data
 GROUP BY city
 ORDER BY avg_aqi DESC limit 10""", db_conn.raw_connection())
# Display the result (Note - the result 'avg_air_quality' is an instance of 'pandas DataFrame
display(first_ten_high)
```

##### <H5>Data Exploration Case 4</H5>   <body>     <p> Find the first ten cities with the lowest average air quality</p>     </body>

```python
first_ten_low = pd.read_sql("""SELECT AVG(AQI) as avg_aqi,
 CityName as city, StateName as state FROM data
 GROUP BY city
 ORDER BY avg_aqi limit 10""", db_conn.raw_connection())
# Display the result (Note - the result 'avg_air_quality' is an instance of 'pandas DataFrame
display(first_ten_low)
```

##### <H5>Data Exploration Case 5</H5>   <body>     <p> Find the average air quality index for each city in Wisconsin </p>     </body>

```python
aqi_city_wisconsin = pd.read_sql("""SELECT AVG(AQI) as avg_aqi,
 CityName as city FROM data
 WHERE StateName = "Wisconsin"
 GROUP BY city
 ORDER BY avg_aqi""", db_conn.raw_connection())
# Display the result (Note - the result 'avg_air_quality' is an instance of 'pandas DataFrame
display(aqi_city_wisconsin)
```

##### <H5>Data Exploration Case 6</H5>   <body>     <p> Find air quality index city Madison and New York ordered by the local date </p>     </body>

```python
aqi_in_madison = pd.read_sql("""SELECT AVG(AQI) as avg_aqi, SUBSTR(DateLocal,0,5) as year, 
 CityName as city FROM data
 WHERE CityName = "Madison"
 GROUP BY SUBSTR(DateLocal,0,5)
""", db_conn.raw_connection())

aqi_in_newyork = pd.read_sql("""SELECT AVG(AQI) as avg_aqi, SUBSTR(DateLocal,0,5) as year, 
 CityName as city FROM data
 WHERE CityName = "New York"
 GROUP BY SUBSTR(DateLocal,0,5)
""", db_conn.raw_connection())
# Display the result (Note - the result 'avg_air_quality' is an instance of 'pandas DataFrame
display(aqi_in_madison)
display(aqi_in_newyork)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
reg = linear_model.Ridge (alpha = .5)
# this is our training label, the AQI of madison - obtained by flattening the DataFrame object
aqis_list_m = list(aqi_in_madison['avg_aqi'].values.flatten())
#aqis_list_n = list(aqi_in_newyork['avg_aqi'].values.flatten())
# This is our training feature 
cities = list(aqi_in_madison['year'].values.flatten())
x_pos = map(int, cities)
x_pos = np.reshape(x_pos, (len(x_pos), 1))


# data
diabetes_X_train = x_pos


# targets
diabetes_y_train = aqis_list_m

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(diabetes_X_train, diabetes_y_train)

# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print("Mean squared error: %.2f"
      % np.mean((regr.predict(diabetes_X_train) - diabetes_y_train) ** 2))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(diabetes_X_train, diabetes_y_train))

# Plot outputs
plt.scatter(diabetes_X_train, diabetes_y_train,  color='black')
plt.plot(diabetes_X_train, regr.predict(diabetes_X_train), color='blue',
         linewidth=3)

plt.xticks(())
plt.yticks(())
plt.savefig("fig2.png")
plt.show()

x_predict = np.array([2017, 2018])
x_predict = np.reshape(x_predict, (len(x_predict), 1))

y_predict = regr.predict(x_predict)
print("The predicted AQI for madison in year 2017 and 2018 are as follows.")
print(y_predict)
```
